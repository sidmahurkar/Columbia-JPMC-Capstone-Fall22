```python
orderedFolders = ["6300_6400_label0","3000_3100_label1","5900_6000_label2","700_800_label3","2400_2500_label4","4000_4100_label5",
           "3600_3700_label6","3300_3400_label7","5000_5100_label8","1800_1900_label9","100_200_label10","0_100_label11"]
```

```python
# use glob to get all the pt files in the folder
path = os.getcwd()
li=[]

for i in range(len(orderedFolders)):
  for value in os.listdir():
    if(orderedFolders[i]==value):
      pt_files = glob.glob(os.path.join(path+"/"+orderedFolders[i]+"/", "*.pt"))
  
      # loop over the list of pt files
      for f in pt_files:
        if((f.split("_")[-1]).split(".")[0]==str(layer_number)):
          li.append(f)
          dic[f]=i;

li = sorted(li)
print(li)
print(len(li))
```
